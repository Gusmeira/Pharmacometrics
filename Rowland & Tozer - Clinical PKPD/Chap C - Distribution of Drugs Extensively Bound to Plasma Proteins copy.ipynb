Explain the apparent differences in tolbutamide disposition between these two groups. What is the cause of the shorter half-life in the subjects with acute viral hepatitis?

_TABLE C-5 | Effect of Acute Viral Hepatitis on Tolbutamide Disposition_

| Subjects              | Half-life (hr) | Volume of Distribution (L/kg) | Clearance (mL/hr/kg) | Fraction of Drug in Plasma Unbound |
|-----------------------|----------------|--------------------------------|----------------------|------------------------------------|
| Healthy               | 5.8            | 0.15                           | 18                   | 0.06                               |
| Acute viral hepatitis | 4.0            | 0.15                           | 26                   | 0.10                               |

In acute viral hepatitis, the fraction of unbound drug increases from 6% to 10%. This could be due to reduced plasma protein binding (e.g., decreased albumin levels or altered protein affinity caused by the disease). Since only the unbound fraction of a drug is available for metabolism and elimination, an increased unbound fraction leads to a higher rate of drug clearance.  
  
Acute viral hepatitis may lead to changes in hepatic enzyme activity. Although hepatitis often reduces liver function, in some cases, enzyme activity related to certain drugs like tolbutamide might be preserved or even increased in early or mild stages of the disease, leading to faster metabolism and clearance.
